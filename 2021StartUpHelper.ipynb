### import rankings from CSV

```python
RB = pd.read_csv('{}RB-{}.csv'.format(data_in,data_postfix))
WR = pd.read_csv('{}WR-{}.csv'.format(data_in,data_postfix))
QB = pd.read_csv('{}QB-{}.csv'.format(data_in,data_postfix))
TE = pd.read_csv('{}TE-{}.csv'.format(data_in,data_postfix))
ALL = pd.read_csv('{}All-{}.csv'.format(data_in,data_postfix))
```

### Import drafted players from sleeper API

```python
draft=requests.get("https://api.sleeper.app/v1/draft/692104825852223488/picks")
draft_df=pd.json_normalize(draft.json())
draft_df['player_name']=draft_df['metadata.first_name'] + ' ' + draft_df['metadata.last_name']
draftedplayers=draft_df['player_name'].replace(name_dict).values
#draftedplayers
```

```python
len(draftedplayers)
```

## Draft Players Function

```python
def draftPlayer(name):
    if(name not in ALL['Full Name'].values):
        print(f'player {name} not in database')
        return
    idx=list(ALL.index[ALL['Full Name']==name])[0]
    ALL.loc[idx,'drafted']=True
    pos=ALL.loc[idx,'Position']
    
    if(pos=='QB'):
        idx_p=list(QB.index[QB['Full Name']==name])[0]
        QB.loc[idx_p,'drafted']=True  
    if(pos=='RB'):
        idx_p=list(RB.index[RB['Full Name']==name])[0]
        RB.loc[idx_p,'drafted']=True  
    if(pos=='WR'):
        idx_p=list(WR.index[WR['Full Name']==name])[0]
        WR.loc[idx_p,'drafted']=True  
    if(pos=='TE'):
        idx_p=list(TE.index[TE['Full Name']==name])[0]
        TE.loc[idx_p,'drafted']=True  
```

```python
for player in draftedplayers:
    draftPlayer(player)
```

## Functions to visualize available players

### define functions

```python
def highlight(s):
    if s.drafted ==True:
        return ['background-color: red']*len(s)
    else:
        return ['background-color: white']*len(s)
```

```python
def print_ALL():
    first=list(ALL.index[ALL.drafted==False])[0]-4
    if(first<0):
        first=0
    if(True in ALL.drafted.values):
        last=list(ALL.iloc[first:first+16].index[ALL.drafted.iloc[first:first+16]==True])[-1]+4
    else:
        last=first+6    
    return ALL.iloc[first:last].style.apply(highlight, axis=1)
```

```python
def print_QB():
    first=list(QB.index[QB.drafted==False])[0]-4
    if(first<0):
        first=0
    if(True in QB.drafted.values):
        last=list(QB.iloc[first:first+16].index[QB.drafted.iloc[first:first+16]==True])[-1]+4
    else:
        last=first+6    
        
    id_all=list(ALL.index[ALL.drafted==False])[0]
    id_qb=list(QB.index[QB.drafted==False])[0]
    points = ALL.iloc[id_all]['Lifetime Value']-QB.iloc[id_qb]['Lifetime Value']
    print(f"{QB.iloc[id_qb]['Full Name']} is {points} lifetime value points behind best available {ALL.iloc[id_all]['Full Name']}" )
    
    return QB.iloc[first:last].style.apply(highlight, axis=1)
```

## Trade Evaluator

```python
def trade_evaluator(A,B):
    A = [round_2_pick(x) for x in A]
    B = [round_2_pick(x) for x in B]
    a_val=ALL.iloc[A]['Lifetime Value'].values.sum()
    b_val=ALL.iloc[B]['Lifetime Value'].values.sum()
    print(f'''Side A has a value of {a_val} \nSide B has a value of {b_val} \nThat is a difference of {a_val-b_val}''')
    
```

```python
def round_2_pick(x):
    round=(x//1-1)
    pick= int((x%1)*100+.01)
    return round*12+pick-1
```
